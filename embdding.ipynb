```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=["the glass of milk",
      "the glass of juice",
      "the cup of tea",
      "i am a good developer",
      "understand the meaning of the words",
      "your vidios are good"]
```

```python
#vocabary siz
voc_size=10000
```

```python
one_hot_repres=[one_hot(words,voc_size)for words in sent] 
one_hot_repres
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
Embedding_doc=pad_sequences(one_hot_repres,padding='pre',maxlen=sent_length)#pre=we are saying the the rest of the size should be covered 
print(Embedding_doc)
#"pre" will write the "0"s at the first and "Post" will write the "0"s at the last
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
model.predict(Embedding_doc)
```
